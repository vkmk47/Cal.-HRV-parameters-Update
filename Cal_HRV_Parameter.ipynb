```python
#Die Methode "find_Time_index" liefert der Index einer gegebenen Zeit zurck
#@param zeit soll genau sein d.h in der list "time" vorhandeln sein ZB. 1.0003987666666667 fr 1min, 5.000398833333334 fr 5min,
#10.000398833333332 fr 10min usw 
#um die genau zeit abzulesen soll die folgende kommenzierter code ausfhrt werden. 
#for t in time:
     #print(t)
    
def find_Time_index(zeit):
    index=0
    for times in time:
        index=index+1
        if(zeit==times):
            break
            
    return index-1
```
